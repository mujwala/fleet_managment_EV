```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
df = pd.read_csv("synthetic_ev_dataset_with_updated_range.csv")

# Plot 1: Bar plot for 'Acceleration 0 - 100 km/h' (mean values per Make)
acceleration_avg = df.groupby('Make')['Acceleration 0 - 100 km/h'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Make', y='Acceleration 0 - 100 km/h', data=acceleration_avg, palette='Blues_d')
plt.title('Average Acceleration (0 - 100 km/h) by Make')
plt.xlabel('Make')
plt.ylabel('Acceleration (s)')
plt.xticks(rotation=45)
plt.show()

# Plot 2: Bar plot for total number of vehicles per Make
plt.figure(figsize=(10, 6))
df['Make'].value_counts().plot(kind='bar', color='orange')
plt.title('Number of Vehicles per Make')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Plot 4: Bar plot for Battery Capacity (BC) by Make
plt.figure(figsize=(10, 6))
bc_avg = df.groupby('Make')['BC'].mean().reset_index()
sns.barplot(x='Make', y='BC', data=bc_avg, palette='Set3')
plt.title('Average Battery Capacity (BC) by Make')
plt.xlabel('Make')
plt.ylabel('Battery Capacity (kWh)')
plt.xticks(rotation=45)
plt.show()

# Plot 5: Bar plot for Monthly Charging Costs by Make
monthly_costs = df.groupby(['Make', 'Month'])[['Charging Cost']].sum().reset_index()
monthly_avg = monthly_costs.groupby('Make')['Charging Cost'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Make', y='Charging Cost', data=monthly_avg, palette='tab10')
plt.title('Average Monthly Charging Costs by Make')
plt.xlabel('Make')
plt.ylabel('Charging Cost ($)')
plt.xticks(rotation=45)
plt.show()

# Additional Plot 6: Bar plot for Gross Vehicle Weight (GVWR)
plt.figure(figsize=(8, 6))
gvwr_avg = df.groupby('Make')['Gross Vehicle Weight (GVWR)'].mean().reset_index()
sns.barplot(x='Make', y='Gross Vehicle Weight (GVWR)', data=gvwr_avg, palette='coolwarm')
plt.title('Average Gross Vehicle Weight (GVWR) by Make')
plt.xlabel('Make')
plt.ylabel('Gross Vehicle Weight (kg)')
plt.xticks(rotation=45)
plt.show()

# Additional Plot 7: Bar plot for Total Power by Drive Type
plt.figure(figsize=(10, 6))
power_avg = df.groupby('Drive')['Total Power'].mean().reset_index()
sns.barplot(x='Drive', y='Total Power', data=power_avg, palette='viridis')
plt.title('Average Total Power by Drive Type')
plt.xlabel('Drive Type')
plt.ylabel('Total Power (kW)')
plt.show()

# Additional Plot 9: Bar plot for Cargo Volume
plt.figure(figsize=(10, 6))
cargo_avg = df.groupby('Make')['Cargo Volume'].mean().reset_index()
sns.barplot(x='Make', y='Cargo Volume', data=cargo_avg, palette='magma')
plt.title('Average Cargo Volume by Make')
plt.xlabel('Make')
plt.ylabel('Cargo Volume (L)')
plt.xticks(rotation=45)
plt.show()

# Additional Plot 10: Count plot for Number of Seats
plt.figure(figsize=(8, 6))
sns.countplot(x='Seats', data=df, palette='pastel')
plt.title('Count of Vehicles by Number of Seats')
plt.xlabel('Number of Seats')
plt.ylabel('Count')
plt.show()

# Additional Plot 11: Bar plot for Maintenance Cost by Make
plt.figure(figsize=(10, 6))
maintenance_avg = df.groupby('Make')['Maintenance Cost'].mean().reset_index()
sns.barplot(x='Make', y='Maintenance Cost', data=maintenance_avg, palette='muted')
plt.title('Average Maintenance Cost by Make')
plt.xlabel('Make')
plt.ylabel('Maintenance Cost ($)')
plt.xticks(rotation=45)
plt.show()


```

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
df = pd.read_csv("synthetic_ev_dataset_with_updated_range.csv")
# Assuming you already have the dataframe `df`

# Group the data by 'Acceleration 0 - 100 km/h' and calculate the mean Top Speed
df_grouped = df.groupby('Acceleration 0 - 100 km/h')['Top Speed'].mean().reset_index()

# Sort the data based on acceleration
df_grouped = df_grouped.sort_values(by='Acceleration 0 - 100 km/h')

# Create the bar plot
plt.figure(figsize=(12, 8))
sns.barplot(data=df_grouped, x='Acceleration 0 - 100 km/h', y='Top Speed', palette='coolwarm')

# Title and labels
plt.title('Top Speed by Acceleration (0-100 km/h)', fontsize=16)
plt.xlabel('Acceleration 0-100 km/h (s)', fontsize=14)
plt.ylabel('Top Speed (km/h)', fontsize=14)

# Customize ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add gridlines for better readability
plt.grid(True, linestyle='--', alpha=0.5)

# Display the plot
plt.show()
# 2. Gross Vehicle Weight (GVWR) vs. Max Payload
plt.figure(figsize=(12, 8))
sns.barplot(data=df, x='Max. Payload', y='Gross Vehicle Weight (GVWR)', errorbar=None, palette='Set2')
plt.title('Gross Vehicle Weight (GVWR) by Max Payload', fontsize=16)
plt.xlabel('Max Payload (kg)', fontsize=14)
plt.ylabel('Gross Vehicle Weight (GVWR) (kg)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# 3. Total Power vs. Total Torque
plt.figure(figsize=(12, 8))
sns.lineplot(data=df.sort_values('Total Power'), x='Total Power', y='Total Torque', marker='o', linewidth=2, color='blue')
plt.title('Relationship between Total Power and Total Torque', fontsize=16)
plt.xlabel('Total Power (kW)', fontsize=14)
plt.ylabel('Total Torque (Nm)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# 4. Wheelbase vs. Cargo Volume
plt.figure(figsize=(12, 8))
sns.barplot(data=df, x='Wheelbase', y='Cargo Volume', errorbar=None, palette='magma')
plt.title('Cargo Volume by Wheelbase', fontsize=16)
plt.xlabel('Wheelbase (mm)', fontsize=14)
plt.ylabel('Cargo Volume (L)', fontsize=14)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# 5. Charging Cost vs. Month
plt.figure(figsize=(12, 8))
monthly_avg = df.groupby('Month')['Charging Cost'].mean().reset_index()
sns.barplot(data=monthly_avg, x='Month', y='Charging Cost', palette='coolwarm')
plt.title('Average Charging Cost by Month', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Charging Cost ($)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming the dataframe is already loaded as `df`
# Create the 'Vehicle Status' column based on 'Range'
df['Vehicle Status'] = df['Range'].apply(
    lambda x: 'Inactive' if x < 250 else 'Active'  # Vehicles with range less than 250 km are considered inactive
)


# 2. Low Range Warning: Distribution of Vehicle Status
df['Low Range Warning'] = df['Range'].apply(lambda x: 'Warning' if x < 250 else 'Normal')

# Bar chart for Low Range Warning distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=df['Low Range Warning'].value_counts().index, y=df['Low Range Warning'].value_counts().values, palette='RdYlGn')
plt.title('Low Range Warning Distribution', fontsize=16)
plt.xlabel('Battery Status', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 3. Performance Comparison: Active vs. Inactive Vehicles
# Bar plot for Top Speed comparison
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Vehicle Status', y='Top Speed', palette='coolwarm')
plt.title('Top Speed Comparison by Vehicle Status', fontsize=16)
plt.xlabel('Vehicle Status', fontsize=14)
plt.ylabel('Top Speed (km/h)', fontsize=14)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Bar plot for Effective Range comparison
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Vehicle Status', y='Range', palette='Blues')
plt.title('Range Comparison by Vehicle Status', fontsize=16)
plt.xlabel('Vehicle Status', fontsize=14)
plt.ylabel('Effective Range (km)', fontsize=14)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 4. Heatmap of vehicle status vs. features
# Create a pivot table for feature averages by vehicle status
status_pivot = df.groupby('Vehicle Status')[['Range', 'Top Speed', 'Length', 'Width', 'Seats']].mean()

plt.figure(figsize=(8, 6))
sns.heatmap(status_pivot, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Vehicle Status vs. Features', fontsize=16)
plt.xlabel('Features', fontsize=14)
plt.ylabel('Vehicle Status', fontsize=14)
plt.tight_layout()
plt.show()
```
